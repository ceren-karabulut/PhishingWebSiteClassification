```
from google.colab import files 


path = "/content/gdrive/MyDrive/CerenPhisingProject/Dataset1.csv"
data = pd.read_csv(path)
```

```
data
```

```
from sklearn.model_selection import train_test_split #train ve test bolumlendirme
```

```
result = data.iloc[:,-1].values
```

```
resultFrame = pd.DataFrame(data=result , index=range(1353), columns=['Result'])
```

```
data2 = data.drop("Result" , axis=1)
```

```
x_train, x_test, y_train , y_test = train_test_split(data2, resultFrame , test_size = 0.30 ,random_state=74 )
```

```
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB,ComplementNB #Naive Bayes 
from sklearn.metrics import accuracy_score #dogruluk orani
from sklearn.metrics import confusion_matrix #confusion matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.model_selection import cross_val_score,cross_validate
```

```
from sklearn.model_selection import StratifiedKFold ,KFold, train_test_split, cross_val_predict, cross_val_score  #Cross Validation
```

```
#Tum algoritmalar için fold iterayonu

skf = StratifiedKFold(n_splits=10)
for train,test in skf.split(data2 , resultFrame):


  x_train , x_test = data2.iloc[train] , data2.iloc[test]
  y_train , y_test = resultFrame.values.ravel()[train] , resultFrame.values.ravel()[test]
```

```
gnb=GaussianNB()
gnb.fit(x_train , y_train)                 #Gaussian Naive Bayes with Cross Validation

y_train_pred = gnb.predict(x_train)
y_test_pred = gnb.predict(x_test)
cm =confusion_matrix(y_test,y_test_pred)

print("Train Classification Report:")
print(classification_report(y_train,y_train_pred ,labels = [-1,0,1]))
print("Train Accuary:%.3f"%accuracy_score(y_train, y_train_pred))
print( "Test Accuary:%.3f"%accuracy_score(y_test,y_test_pred))
print("Classification Report:")
print(classification_report(y_test,y_test_pred ,labels = [-1,0,1]))
print(cm)
```
